### Â§âÊï∞„Å®Âûã

CË®ÄË™û„Å™„Å©„ÅÆ„Éó„É≠„Ç∞„É©„É†Ë®ÄË™û„Åß„ÅØ„ÄÅÂ§âÊï∞„Çí‰Ωø„ÅÜÂ†¥Âêà„ÄÅ‰∫ãÂâç„Å´ÂÆ£Ë®Ä„Åô„ÇãÂøÖË¶Å„Åå„ÅÇ„Çã„ÅÆ„Å†„Åå„ÄÅPython„Åß„ÅØ„ÅÑ„Åç„Å™„Çä„ÄÅÂ§âÊï∞„Å´ÂÄ§„Çí‰ª£ÂÖ•„Åß„Åç„Çã„ÄÇ„Åó„Åü„Åå„Å£„Å¶„ÄÅÂ§âÊï∞„Å´ÂØæ„Åó„Å¶ÂûãÔºàÊï¥Êï∞Âûã„Å®„ÅãÊµÆÂãïÂ∞èÊï∞ÁÇπÂûã„Å®„ÅãÔºâ„ÇíÊòéÁ§∫„Åó„Å™„Åè„Å¶„Çà„ÅÑ„ÄÇ„Åó„Åã„Åó„ÄÅ„Åù„ÅÆÂ§âÊï∞„ÅÆÂûã„ÅØ„ÅÇ„ÇãÁ®ãÂ∫¶ÊÑèË≠ò„Åó„Å¶„ÅÑ„ÇãÂøÖË¶Å„Åå„ÅÇ„Çã„ÄÇ

Â§âÊï∞a„Å´Êï∞ÂÄ§„ÇÑÊñáÂ≠óÂàó„Å™„Å©„Çí‰ª£ÂÖ•„Åó„ÄÅÈñ¢Êï∞print()„Åß„Åù„Çå„ÇíË°®Á§∫„Åô„Çã„ÄÇË®òÂè∑#„ÅØ„Ç≥„É°„É≥„ÉàË°®„Åó„ÄÅ„Åì„ÅÆË®òÂè∑‰ª•Èôç„ÅØ„Ç≥„Éº„Éâ„Å®„Åó„Å¶Ë™çË≠ò„Åï„Çå„Å™„ÅÑ„ÄÇ

````python
# Êï¥Êï∞Âûã
a = 100
print(a)

# ÊµÆÂãïÂ∞èÊï∞ÁÇπÂûã
b = 123.444
print(b)

# ÊñáÂ≠óÂàóÂûã
c = "Hello World"
print(c)

# Ë´ñÁêÜÂûã
d = True
print(d)

# „É™„Çπ„ÉàÂûã
e = [1, 2, 3]
print(e)
````

Ë´ñÁêÜÂûã„ÅØÊï∞ÂÄ§„Å®„Åó„Å¶Êâ±„ÅÜ„Åì„Å®„ÇÇ„Åß„Åç„Çã„ÄÇTrue„ÅØ1„Åß„ÄÅFalse„ÅØ0„Å®„Å™„Çã„ÄÇ‰ª•‰∏ã„ÅÆ‰æã„ÅØÊºîÁÆóÂ≠ê+„Çí‰Ωø„Å£„Å¶„ÄÅTrue„Å®False„ÇíË∂≥„Åó„Å¶„ÅÑ„Çã„Åå„ÄÅÁµêÊûú„ÅØ1„Å®0„ÅÆÂíå„Å®„Å™„Å£„Å¶„ÅÑ„Çã„ÄÇ„Åù„Çå„Å®„ÄÅË®òÂè∑;Ôºà„Çª„Éü„Ç≥„É≠„É≥Ôºâ„Çí‰Ωø„ÅÜ„Å®„ÄÅ1Ë°å„Å´Ë§áÊï∞„ÅÆÂá¶ÁêÜ„ÇíË®òËø∞„Åß„Åç„Çã„ÄÇ

````python
a = True ; b = False
print(a+b)
````

### ÊºîÁÆóÂ≠ê

### „É™„Çπ„Éà

Ë§áÊï∞„ÅÆÂÄ§„Çí„Åæ„Å®„ÇÅ„Å¶Ê†ºÁ¥ç„Åô„Çã„Å´„ÅØ„ÄÅ„É™„Çπ„Éà„Çí‰Ωø„ÅÜ„ÄÇ„É™„Çπ„Éà„Åß„ÅØ„ÄÅË¶ÅÁ¥†„ÅÆÂâçÂæå„Çí[ ]„ÅßÂõ≤„Åø„ÄÅÂêÑË¶ÅÁ¥†„Çí , „ÅßÂå∫Âàá„Çã„ÄÇPython„ÅÆ„É™„Çπ„Éà„Å´„ÅØ„ÄÅ„Å©„ÅÆ„Çà„ÅÜ„Å™Âûã„ÅÆÂÄ§„Åß„ÇÇÊ†ºÁ¥ç„Åß„Åç„Çã„ÄÇ

„É™„Çπ„Éà„ÅÆË¶ÅÁ¥†„Å∏„ÅÆ„Ç¢„ÇØ„Çª„Çπ„Å´„ÅØ„ÄÅÂÖàÈ†≠„Åã„Çâ0, 1, 2, ...„ÅÆ„Çà„ÅÜ„Å™„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ„Çí‰Ωø„ÅÜ„ÄÇ0„Åã„ÇâÂßã„Åæ„Å£„Å¶„ÅÑ„Çã„Åì„Å®„Å´Ê≥®ÊÑè„Åô„Çã„ÄÇ„Åì„ÅÆ„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ„Å´„Çà„Å£„Å¶„ÄÅË¶ÅÁ¥†„ÅÆËøΩÂä†„ÇÑÂÖ•„ÇåÊõø„Åà„ÅåÂèØËÉΩ„Å´„Å™„Çã„ÄÇ

````python
# „É™„Çπ„Éà„Çí‰ΩúÊàê
a = [1, 2, 3, 4, 5]
print(a)

# „Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ3„ÅÆË¶ÅÁ¥†„ÇíÂèñ„ÇäÂá∫„Åô
b = a[3]
print(b)

# Êú´Â∞æ„Å´Ë¶ÅÁ¥†„ÇíËøΩÂä†
a.append(6)
print(a)

# Ë¶ÅÁ¥†„ÇíÂÖ•„ÇåÊõø„Åà„Çã
a[2] = "99"
print(a)
````

„É™„Çπ„Éà„ÅÆË¶ÅÁ¥†„Çí„ÄÅ‰∏ÄÂ∫¶„Å´Â§âÊï∞„Å´‰ª£ÂÖ•„Åô„Çã„ÄÇ

### ifÊñá

Âá¶ÁêÜ„ÅÆÂàÜÂ≤ê„ÇíË°å„ÅÜ„Å´„ÅØ„ÄÅifÊñá„Çí‰Ωø„ÅÜ„ÄÇif„ÅÆÂè≥ÂÅ¥„ÅÆÊù°‰ª∂„ÇíÊ∫ÄË∂≥„Åó„Å™„ÅÑ„Å®„Åç„ÄÅelif„ÅÆÂè≥ÂÅ¥„ÅÆÊù°‰ª∂„ÅåÂà§ÂÆö„Å´‰Ωø„Çè„Çå„Çã„ÄÇif„Å®elif„ÅÆÊù°‰ª∂„ÇíÂÖ®„Å¶Ê∫ÄË∂≥„Åó„Å´„Å®„Åç„ÄÅelseÂÜÖ„ÅÆ„Éñ„É≠„ÉÉ„ÇØ„ÅÆÂá¶ÁêÜ„ÅåÂÆüË°å„Åï„Çå„Çã„ÄÇ

Python„Åß„ÅØ„ÄÅÂàÜÂ≤êÂæå„ÅÆÂá¶ÁêÜ„ÇíË°®„Åô„Éñ„É≠„ÉÉ„ÇØ„ÅÆÁØÑÂõ≤„Çí„ÄÅ„Ç§„É≥„Éá„É≥„Éà„ÅßÁ§∫„Åô„ÄÇ„Ç§„É≥„Éá„É≥„Éà„Åï„Çå„Å¶„ÅÑ„Å™„ÅÑË°å„ÅåÂá∫Áèæ„Åó„Åü„Çâ„ÄÅ„Åù„ÅÆÁõ¥Ââç„ÅÆË°å„Åß„Åù„ÅÆ„Éñ„É≠„ÉÉ„ÇØ„ÅØÁµÇ‰∫Ü„Åó„Å¶„ÅÑ„Çã„Åì„Å®„Å´„Å™„Çã„ÄÇ„Ç§„É≥„Éá„É≥„Éà„Å®„Åó„Å¶„ÄÅÂçäËßí„Çπ„Éö„Éº„Çπ4„Å§„Åå„Çà„Åè‰Ωø„Çè„Çå„Çã„ÄÇ

````python
a = 11
if a < 10:
  print("Good morning!")
elif a < 15:
  print("Good afternoon!")
elif a < 19:
  print("Good evening!")
else:
  print("Good night!")
````

### for Êñá

Áπ∞„ÇäËøî„Åó„ÅÆÂá¶ÁêÜ„ÇíË°å„ÅÜ„Å´„ÅØ„ÄÅforÊñá„Çí‰Ωø„ÅÜ„ÄÇÁπ∞„ÇäËøî„Åó„ÅÆÁØÑÂõ≤„ÇíÊåáÂÆö„Åô„Çã„Åü„ÇÅ„Å´„ÄÅ„É™„Çπ„Éà„ÇÑrangeÈñ¢Êï∞„ÇÑinÊºîÁÆóÂ≠ê„Åå„Çà„Åè‰Ωø„Çè„Çå„Çã„ÄÇ

‰ª•‰∏ã„ÅØrange()„ÅÆ‰Ωø„ÅÑÊñπ„Åß„ÅÇ„Çã„ÄÇÂºïÊï∞„ÅÆ[ ]„ÅßÂõ≤„Åæ„Çå„ÅüÁÆáÊâÄ„ÅØÁúÅÁï•„Åß„Åç„Çã„ÄÇ

range([ÈñãÂßãÁï™Âè∑,] ÁµÇ‰∫ÜÁï™Âè∑ [, „Çπ„ÉÜ„ÉÉ„ÉóÊï∞])

````python
# „É™„Çπ„Éà„Çí‰Ωø„Å£„ÅüforÊñá
for a in [3, 5, 9, 17]:
  print(a)

# range()„Çí‰Ωø„Å£„ÅüforÊñá
for b in range(6):
  print(b)
````

### csv„Éá„Éº„Çø„ÅÆË™≠„ÅøÊõ∏„Åç

DataFrame„ÅØ„ÄÅÊ¨°„ÅÆ„Çà„ÅÜ„Å´ÂÆöÁæ©„Åô„Çã„ÄÇ

pd.DataFrame(data = Ê†ºÁ¥ç„Åô„Çã„Éá„Éº„Çø, index = Ë°åÂêç, columns = ÂàóÂêç)

 * data
  	- DataFrame„Å´Ê†ºÁ¥ç„Åô„Çã„Éá„Éº„Çø„ÇíÈÖçÂàó„Å™„Å©„ÅßÊåáÂÆö
 * index
   -	ÂêÑË°å„ÅÆË°åÂêçÔºà„Ç§„É≥„Éá„ÉÉ„ÇØ„ÇπÔºâ„ÇíÈÖçÂàó„Å™„Å©„ÅßÊåáÂÆö
   - ÊåáÂÆö„Åó„Å™„Åë„Çå„Å∞„ÄÅÔºëË°åÁõÆ„Åã„ÇâÈ†Ü„Å´0,1,2„Å®ÈÄ£Áï™
 * columns
   - ÂêÑÂàó„ÅÆÂàóÂêç„ÇíÈÖçÂàó„Å™„Å©„ÅßÊåáÂÆö

````python
import numpy as np
arr1 = np.array([[11,12,13], [21,22,23], [31,32,33]])

import pandas as pd
index1 = ["Row1", "Row2", "Row3"]
columns1 =["Col1", "Col2", "Col3"]
df3 = pd.DataFrame(data=arr1, index=index1, columns=columns1)
print(df3)
````

‰ª•‰∏ã„ÅÆ„Ç≥„Éº„Éâ„ÅØ„ÄÅpyplot„Çí‰Ωø„Å£„Å¶Áõ¥Á∑ö„ÇíÊèèÁîª„Åô„Çã„ÄÇ„Åæ„Åö„ÄÅNumPy„ÅÆlinspace()Èñ¢Êï∞„Çí‰Ωø„Å£„Å¶„ÄÅxÂ∫ßÊ®ô„ÅÆ„Éá„Éº„Çø„ÇíÈÖçÂàó„Å®„Åó„Å¶ÁîüÊàê„Åô„Çã„ÄÇ„Åì„ÅÆxÂ∫ßÊ®ô„ÅÆ„Éá„Éº„Çø„Å´Êï∞ÂÄ§„Çí„Åã„Åë„Åü„ÇÇ„ÅÆ„ÇíyÂ∫ßÊ®ô„Å´Ë®≠ÂÆö„Åô„Çã„ÄÇ„Åù„ÅÆ‰∏ä„Åß„ÄÅpyplot„ÅÆplot„Çí‰Ωø„Å£„Å¶„ÄÅxÂ∫ßÊ®ô„Åä„Çà„Å≥yÂ∫ßÊ®ô„ÅÆÂÄ§„Çí„Éó„É≠„ÉÉ„Éà„Åó„ÄÅshow„Åß„Ç∞„É©„Éï„Å´Ë°®Á§∫„Åô„Çã„ÄÇ

````python
import numpy as np
import matplotlib.pyplot as plt

j1 = np.linspace(-5, 5, num = 20)   # -5„Åã„Çâ5„Åæ„Åß„ÅÆÂå∫Èñì„Çí19Á≠âÂàÜ„Åô„Çã
j2 = 2 * j1   # j1„Å´2„Çí„Åã„Åë„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã

plt.plot(j1, j2)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

j1 = np.linspace(-5, 5, num = 20)   # -5„Åã„Çâ5„Åæ„Åß„ÅÆÂå∫Èñì„Çí19Á≠âÂàÜ„Åô„Çã
j2 = 9 * j1   # j1„Å´2„Çí„Åã„Åë„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã

plt.plot(j1, j2)
plt.show()
````

### „Ç∞„É©„Éï„ÅÆË°®Á§∫Ë®≠ÂÆö

„Ç∞„É©„Éï„ÅÆËª∏„ÅÆ„É©„Éô„É´„ÇÑ„Çø„Ç§„Éà„É´„ÄÅÂá°‰æã„Å™„Å©„ÇíË°®Á§∫„Åó„ÄÅÁ∑ö„ÅÆ„Çπ„Çø„Ç§„É´„ÇíÂ§âÊõ¥„Åô„ÇãÊñπÊ≥ï„Çí‰æãÁ§∫„Åô„Çã„ÄÇ

````python
import numpy as np
import matplotlib.pyplot as plt

j1 = np.linspace(-5, 5, num = 20)   # -5„Åã„Çâ5„Åæ„Åß„ÅÆÂå∫Èñì„Çí19Á≠âÂàÜ„Åô„Çã
j2 = 3 * j1   # j1„Å´3„Çí„Åã„Åë„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã
j3 = j1**2    # j1„Çí2‰πó„Åó„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã

# Ëª∏„ÅÆ„É©„Éô„É´
plt.xlabel("x axis")
plt.ylabel("y axis")

# „Ç∞„É©„Éï„ÅÆ„Çø„Ç§„Éà„É´
plt.title("Graph Setting")

# „Éó„É≠„ÉÉ„Éà„ÄÅÂá°‰æã„Å®Á∑ö„ÅÆ„Çπ„Çø„Ç§„É´„ÇíË®≠ÂÆö
plt.plot(j1, j2, label="x * 3", color="red")
plt.plot(j1, j3, label="x**2", linestyle="dashed")

plt.xlim(-5, 5)
plt.ylim(-10, 20)
plt.legend()   # Âá°‰æã

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

j1 = np.linspace(20, 40, num = 20)   # -5„Åã„Çâ5„Åæ„Åß„ÅÆÂå∫Èñì„Çí19Á≠âÂàÜ„Åô„Çã
j2 = 5 * j1   # j1„Å´3„Çí„Åã„Åë„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã
j3 = j1**2    # j1„Çí2‰πó„Åó„Å¶yÂ∫ßÊ®ô„Å®„Åô„Çã

# Ëª∏„ÅÆ„É©„Éô„É´
plt.xlabel("x axis")
plt.ylabel("y axis")

# „Ç∞„É©„Éï„ÅÆ„Çø„Ç§„Éà„É´
plt.title("Graph Setting")

# „Éó„É≠„ÉÉ„Éà„ÄÅÂá°‰æã„Å®Á∑ö„ÅÆ„Çπ„Çø„Ç§„É´„ÇíË®≠ÂÆö
plt.plot(j1, j2, label="x * 3", color="red")
plt.plot(j1, j3, label="x**2", linestyle="dashed")

plt.xlim(20, 40)
plt.ylim(-10, 15)
plt.legend()   # Âá°‰æã

plt.show()
````

### Êï£Â∏ÉÂõ≥

Êï£Â∏ÉÂõ≥„ÇíË°®Á§∫„Åô„Çã„Å´„ÅØ„ÄÅpyplot„ÅÆscatter()Èñ¢Êï∞„Çí‰Ωø„ÅÜ„ÄÇ‰ª•‰∏ã„ÅÆ„Ç≥„Éº„Éâ„ÅØ„ÄÅxÂ∫ßÊ®ô„ÄÅyÂ∫ßÊ®ô„ÅÆÈÖçÂàó„Å´Âü∫„Å•„ÅÑ„Å¶Êï£Â∏ÉÂõ≥„ÇíÊèèÁîª„Åô„Çã„ÄÇ

scatter()Èñ¢Êï∞„ÅÆÂºïÊï∞„Å´„ÅØ„ÄÅ‰ª•‰∏ã„Åå„ÅÇ„Çã„ÄÇ
*   sÔºö„Çµ„Ç§„Ç∫Ôºàdefault„ÅØ20Ôºâ
*   cÔºöËâ≤
*   alphaÔºöÈÄèÊòéÂ∫¶Ôºà0ÔΩû1„ÅÆÊï∞ÂÄ§„ÇíÊåáÂÆöÔºâ
*   linewidthsÔºöÁ∑ö„ÅÆÂ§™„Åï
*   edgecolorsÔºöÁ∑ö„ÅÆËâ≤


````python
import numpy as np
import matplotlib.pyplot as plt

j4 = np.random.rand(10) + 1.0   # 0ÔΩû1„ÅÆ‰π±Êï∞„Çí10ÂÄãÁîüÊàê„Åó„ÄÅ1„ÇíË∂≥„Åô
j5 = np.random.rand(10) + 1.0

# Ëª∏„ÅÆ„É©„Éô„É´
plt.xlabel("x axis")
plt.ylabel("y axis")

# Êï£Â∏ÉÂõ≥„Çí„Éó„É≠„ÉÉ„Éà
plt.scatter(j4, j5, s=600, c="pink", alpha=0.5, linewidths=2, edgecolors="red")

# Ëª∏„ÅÆÁØÑÂõ≤„ÇíÊåáÂÆö
plt.xlim(1, 2)
plt.ylim(1, 2)

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

j4 = np.random.rand(1000) + 1   # 0ÔΩû1„ÅÆ‰π±Êï∞„ÇínÂÄãÁîüÊàê„Åó„ÄÅ1„ÇíË∂≥„Åô
j5 = np.random.rand(1000) + 1

# Ëª∏„ÅÆ„É©„Éô„É´
plt.xlabel("xËª∏„Åò„Åè„É©„Éô„É´„Çìn")
plt.ylabel("yËª∏„Åò„ÅèR„É©„Éô„É´„Çìn")

# Êï£Â∏ÉÂõ≥„Çí„Éó„É≠„ÉÉ„Éà
#scatter()Èñ¢Êï∞„ÅÆÂºïÊï∞„Å´„ÅØ„ÄÅ‰ª•‰∏ã„Åå„ÅÇ„Çã„ÄÇ

#sÔºö„Çµ„Ç§„Ç∫Ôºàdefault„ÅØ20Ôºâ,cÔºöËâ≤,alphaÔºöÈÄèÊòéÂ∫¶Ôºà0ÔΩû1„ÅÆÊï∞ÂÄ§„ÇíÊåáÂÆöÔºâ
#linewidthsÔºöÁ∑ö„ÅÆÂ§™„Åï,edgecolorsÔºöÁ∑ö„ÅÆËâ≤
plt.scatter(j4, j5, s=10, c="red", alpha=0.3, linewidths=0.3, edgecolors="blue")

# Ëª∏„ÅÆÁØÑÂõ≤„ÇíÊåáÂÆö
plt.xlim(1, 2)
plt.ylim(1, 2)

plt.show()
````
