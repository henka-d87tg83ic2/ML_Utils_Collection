```python
# 必要なライブラリをインポート
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro
import statsmodels.api as sm
from google.colab import files   # Colab用のファイルアップロード
import logging
from datetime import datetime
from IPython.display import HTML  # ColabでHTML表示用

# ロギング設定
logging.basicConfig(level=logging.INFO)

# ファイルアップロード関数
def upload_file():
    uploaded = files.upload()  # ファイルアップロードダイアログを表示
    for file_name in uploaded.keys():
        print(f"アップロードされたファイル: {file_name}")
        return file_name

# CSVファイル読み込み関数
def load_csv(file_name):
    """CSVファイルを読み込み、データフレームを返す"""
    df = pd.read_csv(file_name)
    print("データの確認（先頭5行）:")
    # HTML形式で表示
    display(HTML(df.head().to_html()))
    return df

# 実行例
try:
    logging.info("ファイルアップロードを開始")
    file_name = upload_file()  # ファイルアップロードを実行
    if file_name:
        logging.info(f"ファイル '{file_name}' がアップロードされました")

        # CSVファイルを読み込む
        df = load_csv(file_name)
except Exception as e:
    logging.error(f"エラーが発生しました: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import io
import base64
from IPython.display import HTML

# CSVファイルを読み込む
df3 = pd.read_csv("sample-data1.csv")

# weight列とpressure列を取得
x1 = df3["weight"]
y1 = df3["pressure"]

# 散布図を描画
plt.figure()
plt.xlabel("weight")
plt.ylabel("pressure")
plt.scatter(x1, y1, s=100, c="blue")

# 画像をメモリに保存してBase64に変換
buf = io.BytesIO()
plt.savefig(buf, format="png")
buf.seek(0)
img_base64 = base64.b64encode(buf.read()).decode("utf-8")
buf.close()

# matplotlibの自動表示を抑制
plt.close()

# HTMLとして表示
html_code = f'<img src="data:image/png;base64,{img_base64}" />'
display(HTML(html_code))
```
