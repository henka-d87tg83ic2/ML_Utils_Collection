# ニューラルネットワーク

irisデータセット

```python
from sklearn.datasets import load_iris

iris = load_iris() #irisデータをロード

from sklearn.model_selection import train_test_split
#データを訓練用と試験用に分割
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=1, shuffle = True)

# StandardScalerを使って、平均0分散1になるようにスケーリング
from sklearn.preprocessing import StandardScaler
# 特徴量（入力データ）
ss = StandardScaler() # 標準化のためのスケーラー
ss.fit(x_train) # 変換のためのパラメータ（平均と標準偏差）を計算
x_train_std = ss.transform(x_train) # データの変換（標準化の実行）
x_test_std = ss.transform(x_test)
print(x_train_std[:5]) # 入力でデータを5件表示
print(x_test_std[:5])
```

### モデルの用意

```python
from sklearn.neural_network import MLPClassifier
# Define Neural Neowork model
clf = MLPClassifier(hidden_layer_sizes=10, activation='relu', solver='sgd', max_iter=10000)
```

### モデルの学習と分類精度

```python
# Lerning model
clf.fit(x_train_std, y_train)

# Predict test data
y_pred = clf.predict(x_test_std)

# Calculate prediction accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))

```

### 予測

モデルによって試験用データを分類した結果と正解ラベルとを比較する。

```python
import numpy as np
print(y_pred)      # 予測結果を表示
print(y_test)      # 正解ラベルを表示
```

学習過程の表示

```python
import matplotlib.pyplot as plt
# Show loss curve
plt.plot(clf.loss_curve_)
plt.title("Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid()
plt.show()

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

loss_curve = clf.loss_curve_

# 1. 折れ線グラフ
plt.figure(figsize=(8,5))
plt.plot(loss_curve, color="blue", marker="o")
plt.title("Loss Curve (折れ線グラフ)")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# 2. 散布図
plt.figure(figsize=(8,5))
plt.scatter(range(len(loss_curve)), loss_curve, color="red")
plt.title("Loss Curve (散布図)")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# 3. 移動平均曲線
window = 5
moving_avg = np.convolve(loss_curve, np.ones(window)/window, mode="valid")
plt.figure(figsize=(8,5))
plt.plot(range(window-1, len(loss_curve)), moving_avg, color="green")
plt.title("Loss Curve (移動平均)")
plt.xlabel("Iteration")
plt.ylabel("Loss (Smoothed)")
plt.grid(True)
plt.show()

# 4. ヒートマップ風
plt.figure(figsize=(12,2))
sns.heatmap(np.array(loss_curve).reshape(1,-1), cmap="coolwarm", annot=False,
            xticklabels=range(len(loss_curve)), yticklabels=["Loss"])
plt.title("Loss Curve (ヒートマップ風)")
plt.show()
```
