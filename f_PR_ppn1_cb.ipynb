```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron # Perceptronのインポート
iris = load_iris()
# print(iris)
X1 = iris.data[:,[0,1,2,3]]
y1 = iris.target
model = Perceptron().fit(X1, y1)
model.score(X1, y1)
#-----------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# データロード
iris = load_iris()
X = iris.data
y = iris.target
species = iris.target_names

# モデル学習
model = Perceptron(max_iter=1000, random_state=0)
model.fit(X, y)
score = model.score(X, y)
print("学習データに対するスコア:", score)

# 1. 散布図（花弁の長さと幅）
plt.figure(figsize=(6,5))
sns.scatterplot(x=X[:,2], y=X[:,3], hue=[species[i] for i in y], palette="Set1")
plt.title("花弁の長さと幅による分類可視化")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.show()

# 2. 決定境界の可視化（2次元射影）
X2 = X[:, [2,3]]  # 花弁の長さと幅
model2 = Perceptron(max_iter=1000, random_state=0).fit(X2, y)

x_min, x_max = X2[:,0].min()-0.5, X2[:,0].max()+0.5
y_min, y_max = X2[:,1].min()-0.5, X2[:,1].max()+0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))
Z = model2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(6,5))
plt.contourf(xx, yy, Z, alpha=0.3, cmap="Set1")
sns.scatterplot(x=X2[:,0], y=X2[:,1], hue=[species[i] for i in y], palette="Set1", edgecolor="k")
plt.title("Perceptronによる決定境界")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.show()

# 3. 混同行列
y_pred = model.predict(X)
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=species)
disp.plot(cmap="Blues")
plt.title("混同行列")
plt.show()

# 4. 分類レポート
print("分類レポート:\n", classification_report(y, y_pred, target_names=species))
```

```python
# 訓練データと検証データに分割
from sklearn.model_selection import train_test_split
# 全体の30%を検証データにする
X_train, X_test, y_train, y_test = train_test_split(
    X0, y0, test_size = 0.3, random_state = 1, shuffle = True
)
```

```python
# StandardScalerを使って、平均0分散1になるようにスケーリング
from sklearn.preprocessing import StandardScaler
# インスタンスの作成
sc = StandardScaler()
# 訓練データの平均と標準偏差を計算
sc.fit(X_train)
# 標準化実行（標準化は、訓練データの平均と標準偏差を用いて行う）
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# Perceptronで学習
from sklearn.linear_model import Perceptron
# epoch数50, 学習率0.1で、Perceptronのインスタンス作成
ppn0 = Perceptron(max_iter = 40, eta0 = 0.1, random_state = 2, shuffle = True)
# 訓練データにモデルを適合させる
ppn0.fit(X_train_std, y_train)
```

```python
# 検証データで分類精度を計算
y_pred = ppn0.predict(X_test_std)
# 誤って分類したものの数を表示
print('MissClassified Samples : %d' % (y_test != y_pred).sum())     # sum()は総計
```

```python
# 正解率
from sklearn.metrics import accuracy_score
print('Accuracy : %.2f' % accuracy_score(y_test, y_pred))

#-----------------------
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# 正解率の計算
acc = accuracy_score(y_test, y_pred)
print('Accuracy : %.2f' % acc)

# 1. 棒グラフで正解率を表示
plt.figure(figsize=(4,6))
plt.bar(["Accuracy"], [acc], color="skyblue")
plt.ylim(0,1.1)
plt.title("正解率 (棒グラフ)")
plt.show()

# 2. ゲージ風グラフ
plt.figure(figsize=(6,6))
circle = plt.Circle((0,0), 1, color="lightgray")
fig, ax = plt.subplots(figsize=(6,6))
ax.add_artist(circle)
theta = acc * 360
ax.pie([theta, 360-theta], colors=["skyblue","white"], startangle=90, counterclock=False)
plt.title("正解率 (ゲージ風)")
plt.show()

# 3. 混同行列
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("混同行列")
plt.show()

# 4. クラスごとの精度棒グラフ
class_accuracy = cm.diagonal() / cm.sum(axis=1)
plt.figure(figsize=(6,4))
sns.barplot(x=range(len(class_accuracy)), y=class_accuracy, palette="coolwarm")
plt.ylim(0,1.1)
plt.xlabel("Class Label")
plt.ylabel("Accuracy")
plt.title("クラスごとの正解率")
plt.show()
```
